```python
#export
"let's make write file a function"
def write_fetch_result(filename,content, dirname="tmp_smile_first/"):
    from pathlib import Path
    "save the file"
    "check if the folder exist first"
    if not os.path.isdir(Path(dirname)):
        os.system("mkdir %s" % dirname)
    else:
        print("folder exists")
    "write content into the file"
    if not os.path.isfile(Path(dirname+filename)):
        with open(Path(dirname+filename), "w") as f:
            cnt = 0
            for ele in content:
                f.write("\n\nfile #%d\n"%(cnt))
                f.write(ele)
                cnt += 1
            f.close()        
    else:
        print("is file, skipped")
```

### process of fetching thousands of posts
#### fetch as much as we can at once (30)
#### -> save to files (e.g, file_1.json)
#### -> repeat last 2 steps until we get the amount we want
#### -> merge the existing files to a big ass file 

## Or, we can just append everything to a list and suk it up

```python
#export
"first, make fetch emoji a function as well"
def fetch_emoji(emoji='',plurk=None):
    if len(emoji) == 0:
        print("emoji not inputed, exit")
        return
    if plurk is None:
        print("initialize plurk ...")
        from acc import acc_info
        from plurk_oauth import PlurkAPI
        # with access token, just run this
        plurk = PlurkAPI(acc_info["CONSUMER_KEY"], acc_info["CONSUMER_SECRET"])
        plurk.authorize(acc_info["API_KEY"], acc_info["API_SECRET"])
    tmp_p = plurk.callAPI('/APP/PlurkSearch/search', options={'query' : emoji,'offset':100})
    # weird edge case exists
    if tmp_p is None:
        return []
    # should we check if key exists?
    return [ele['content_raw'] for ele in tmp_p['plurks'] if ele is not None]
```

```python
%%time
big_content = []
" loop thru it"
"say, the limit per post is 30, we want 10000 (for example)"
"so, we should fetch 334 times"

for _ in range(10000//30+1):
    # make this a big list
    # we should add: if fetched list is [] then just break the loop
    big_content += fetch_emoji(emoji='😂',plurk=plurk)

"then, write all the content at once"
write_fetch_result(filename="bigass.json", content=big_content)
```
